```python
#importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#deleting nan values
tdf.dropna()
```

Percentile calculation

```python
#defining a new column having percentile of retweets column
tdf["per"]=tdf.retweets.rank(pct=True)
```

Binning

```python
#dividing data acoording to percentile of the retweets column
#qcut will divie data into 4 bins
#making a new column and allocating the data into 4 categories
#the 4 bins are lowest retweets, low retweets, medium retweets, maximum retweets
#lowest retweets has the data whose percentile is less than 0.9 as although this has the maximum no. of columns but it is not much useful
#low retweets has data whose percentile greater than 0.9 but lesser than 0.99
#as we move forward the variation in percentile is decresing
#medium retweets contains data with percentile greater than 0.99 but lesser than 0.999
#maximum retweets contains data wuth percentile greater than 0.999
#the maximum retweets section is the most crucial section for as it tells about what is trending among the users
#tdf['perqcut'] = pd.qcut(tdf.per, q=[0,0.9,0.99,0.999,1.0], 
 #                       labels=[0.9, 
  #                              0.99, 
   #                             0.999,  
    #                            1.0])
```

```python
#create bins with percentile column of retweets
tdf['per_quantile_bin']=pd.qcut(tdf['per'],q=10,precision=0,duplicates='drop')
```

```python
sns.countplot(tdf['per_quantile_bin'])
```
